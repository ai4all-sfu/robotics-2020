```python
import cv2
import matplotlib.pyplot as plt
import time 
import os
%matplotlib inline 
```

```python
def convertToRGB(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

load cascade classifier training file for haarcascade

```python
cv2_base_dir = os.path.dirname(os.path.abspath(cv2.__file__))
haar_model = os.path.join(cv2_base_dir, 'data/haarcascade_frontalface_default.xml')
haar_face_cascade = cv2.CascadeClassifier(haar_model)
```

Loading image for detecting the faces

```python
test1 = cv2.imread('data/babies.jpg')
```

Convert the test image to gray image as opencv face detector expects gray images

```python
gray_img = cv2.cvtColor(test1, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
```

```python
#let's detect multiscale (some images may be closer to camera than others) images
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5);
```

```python
#print the number of faces found
print('Faces found: ', len(faces))

#go over list of faces and draw them as rectangles on original colored img
for (x, y, w, h) in faces:
    cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
#convert image to RGB and show image
plt.imshow(convertToRGB(test1))
```

```python
haar_model_eye = os.path.join(cv2_base_dir, 'data/haarcascade_eye.xml')
haar_eye_cascade = cv2.CascadeClassifier(haar_model_eye)
eyes = haar_eye_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=5);

#go over list of faces and draw them as rectangles on original colored img
for (x, y, w, h) in eyes:
    cv2.rectangle(test1, (x, y), (x+w, y+h), (0, 255, 0), 2)
    
#convert image to RGB and show image
plt.imshow(convertToRGB(test1))
```
